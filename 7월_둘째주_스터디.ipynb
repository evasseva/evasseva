```python
# 선택정렬

a = [2, 4, 5, 1, 3]

for i in range(len(a)):
    min = i
    for j in range(i+1, len(a)):
        if a[min] > a[j]:
          min=j
    a[i], a[min] = a[min], a[i]


print(a)
```

```python
# 삽입정렬

a = [2, 4, 5, 1, 3]

for i in range(1, len(a)):
  tmp = a[i]
  idx = i
  for j in range((i-1), -1, -1):
    if a[i] < a[j]:
      idx = j
  for k in range(i, idx, -1):
    a[k] = a[k-1]
  a[idx] = tmp

print(a)
```
